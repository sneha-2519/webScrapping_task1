```python
## Web Scraping using Python with url = 'https://www.amazon.com/dp/B0C1J4MPBB/'
```

```python
# Importing required libraries

import requests
from bs4 import BeautifulSoup
import json
```

```python
# Enter URL
URL = 'https://www.amazon.com/dp/B0C1J4MPBB/'
```

```python
# Assign required Headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36",
    "Accept-Encoding": "gzip, deflate",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
    "DNT": "1",
    "Connection": "close",
    "Upgrade-Insecure-Requests": "1"
}
```

```python
from re import T
try:
    page = requests.get(URL, headers=headers) # Requesting fetch the url and headers
    page.raise_for_status()           # Check the status code 200

    soup = BeautifulSoup(page.content, "html.parser") # BeautifulSoup library from Python's bs4 module. prase the html code from web page

    product_data = {} # assign empty Dictonary

    # fetching the Title of product
    Title = soup.find(id='productTitle').get_text().strip()
    if Title:
        soup.find(id='productTitle')
    else:
        "Title not found"
    product_data["Title"] = Title


    # fetching the Price of product
    Price_element = soup.find(class_="a-price-whole")
    if Price_element:
        Price_text = Price_element.get_text().strip()
        product_data["price"] = Price_text
    else:
        product_data["price"] = "Price not found"

    # fetching the rating of product
    Rating_element = soup.find(id="acrPopover")
    if Rating_element:
        rating_text = Rating_element.text.strip()
        product_data["rating"] = rating_text
    else:
        product_data["rating"] = "Rating not found"

    # fetching the No_of_rating of product
    No_of_rating = soup.find(id="acrCustomerReviewText").get_text().strip()
    if No_of_rating:
      soup.find(id="acrCustomerReviewText")
    else:
      print("Number of Ratings not found")
    product_data["number_of_ratings"] = No_of_rating

    # fetching the Image of product
    Image = soup.find(id="imgTagWrapperId")
    if Image:
        image_url = Image.find("img")["src"]
        product_data["image_url"] = image_url
    else:
        product_data["image_url"] = "Image URL not found"

    # fetching the Additional Information of product
    table_ele = soup.find(id="poExpander")
    if table_ele:
        product_details = {}
        for row in table_ele.find_all('tr'):  # Iterate through table rows
            key_cell, value_cell = row.find_all('td')  # Assuming each row has two cells (key, value)
            if key_cell and value_cell:
                key = key_cell.text.strip().replace(':', '')
                value = value_cell.text.strip()
                product_details[key] = value
        product_data["Additional_Information"] = product_details  # Nested dictionary for details
    else:
        product_data["Additional_Information"] = "Table element not found."

    json_data = json.dumps(product_data, indent=4)  # Convert to JSON with indentation for readability
    print(json_data)

except requests.exceptions.RequestException as e:
    print(f"An error occurred while fetching the product data: {e}")
```
